```python
#To create another variable average_score which will be the Average Score of the students, I will first get the total marks per student
data['total score']=data['mathScore']+data['readingScore']+data['writingScore']
```

```python
#Now I will divide the total marks by 3 to get the average score
data['average score']=data['total score']/3
data
```

```python
def category(scores):
    if scores >= 85 and scores <= 100:
        return 'A1'
    elif scores >= 70 and scores < 85:
        return 'B2'
    elif scores >= 65 and scores < 70:
        return 'B3'
    elif scores >= 60 and scores < 65:
        return 'C4'
    elif scores >= 55 and scores < 60:
        return 'C5'
    elif scores >= 50 and scores < 55:
        return 'C6'
    elif scores >= 45 and scores < 50:
        return 'D7'
    elif scores >= 40 and scores < 45:
        return 'E8'
    elif scores >= 0 and scores < 40:
        return 'F9'
    
data['Average Score Category']=data['average score'].apply(category)

data
```

```python
#Graphical representations showing relationships beteween each variable with the average score category

plt.figure(figsize=(15,5))

plt.subplot(1,4,1)
sns.countplot(x = "Average Score Category" ,hue="gender" , data=data ,palette="dark")
plt.xticks(rotation=90)
plt.title("Category Against Gender Type")

plt.subplot(1,4,2)
sns.countplot(x = "Average Score Category" ,hue="group" , data=data ,palette="Greens_d")
plt.xticks(rotation=90)
plt.title("Category Against Ethnicity")

plt.subplot(1,4,3)
sns.countplot(x = "Average Score Category" ,hue="parentEdu" , data=data ,palette="GnBu_d")
plt.xticks(rotation=90)
plt.title("Category Against Parents Education")

plt.subplot(1,4,4)
sns.countplot(x = "Average Score Category" ,hue="lunch" , data=data ,palette="Greens_d")
plt.xticks(rotation=90)
plt.title("Category Against Lunch Type")
plt.show()

#it is seen here that 1. Grade B2 has the highest population between both male and female genders, with the females recording the highest value.
#                     2. Grade B2 has the highest population across all Ethnicity/Groups, with those in group C recording the highest value.
#                     3. Grade B2 has the highest population across all Parents Education Levels, with those, whose parents having "some college degree" recording the highest value.
#                     4. Grade B2 has the highest population across both Lunch packages, with the Standard lunch package recording the higher value.
```
